```python
# import libraries
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# read in dataset and split it to 80% training data and 20% testing data
diabetes = pd.read_csv('./dataset_diabetes/diabetic_data_clean.csv')
train_set, test_set = train_test_split(diabetes, test_size = 0.2)
```
